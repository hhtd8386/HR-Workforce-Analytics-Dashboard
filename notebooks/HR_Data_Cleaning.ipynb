```python
import pandas as pd
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

Project Paths

```python
project_root = Path("..")

input_path = project_root / "data" / "Data.xlsx"
output_path = project_root / "data" / "Data_Cleaned_Final.xlsx"

print(f"Input File : {input_path}")
print(f"Output File: {output_path}")
```

Load Workbook

```python
excel = pd.ExcelFile(input_path)

print(f"Workbook Loaded Successfully")
print(f"Total Sheets: {len(excel.sheet_names)}")

print("\nSheets:")

for sheet in excel.sheet_names:
    print(f"• {sheet}")
```

Cleaning Function

```python
def clean_dataframe(df):

    df.columns = df.columns.str.strip()

    text_cols = df.select_dtypes(include="object").columns

    for col in text_cols:

        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .replace(["", "nan", "None"], pd.NA)
        )

    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    date_cols = [c for c in df.columns if "date" in c.lower()]

    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors="coerce")

    df = df.drop_duplicates()

    return df
```

Validation Function

```python

def validate_dataframe(df):
    print("\nData Validation Report")
    print("-" * 50)
    print(f"Rows            : {len(df)}")
    print(f"Columns         : {len(df.columns)}")
    print(f"Duplicate Rows  : {df.duplicated().sum()}")
    missing = df.isna().sum()
    print("\nMissing Values")
    if missing.sum() == 0:
        print("No missing values detected.")
    else:
        print(missing[missing > 0])
    print("\nData Types")
    display(df.dtypes)
    date_cols = [c for c in df.columns if "date" in c.lower()]
    if date_cols:
        print("\nInvalid Dates")
        for col in date_cols:
            print(f"{col}: {df[col].isna().sum()} invalid values")
    numeric_cols = df.select_dtypes(include="number").columns
    print("\nNegative Values")
    found = False
    for col in numeric_cols:
        negative = (df[col] < 0).sum()
        if negative > 0:
            found = True
            print(f"{col}: {negative}")
    if not found:
        print("No negative values detected.")
```

Process All Sheets

```python
with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    for sheet in excel.sheet_names:
        print("=" * 70)
        print(f"Processing Sheet: {sheet}")
        df = pd.read_excel(excel, sheet_name=sheet)
        rows_before = len(df)
        df = clean_dataframe(df)
        rows_after = len(df)
        print(f"Rows Before : {rows_before}")
        print(f"Rows After  : {rows_after}")
        print(f"Removed     : {rows_before - rows_after}")
        validate_dataframe(df)
        df.to_excel(writer, sheet_name=sheet, index=False)
print("\nCleaning Completed Successfully!")
print(f"Clean dataset saved to:\n{output_path}")
```
